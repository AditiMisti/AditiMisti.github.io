```python
import pandas as pd
```

```python
df = pd.read_csv(
    "TMDB_all_movies.csv",
    on_bad_lines="skip",
    engine="python"
)
```

```python
print(df.head())
```

```python
df = df.drop_duplicates()
```

```python
df = df.dropna(subset=['title'])
```

```python
df = df[['title','overview',
'genres','tagline',
'vote_average','popularity']]
```

```python
df['overview'] = df['overview'].fillna('')
```

```python
df['tagline'] = df['tagline'].fillna(' ')
```
